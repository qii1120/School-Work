```python
import csv
import pandas as pd

datas = pd.read_csv("data.csv")
df = pd.DataFrame(columns = ['High school system', 'Gender', 'Num of graduates', 'Num of enroll in police university', 'Num of enroll in police college',
              "average num of enroll in military school", "Police university rate", "Police college rate", "Military school enrollment rate"])
national = datas[datas["High school system"].str.startswith("National")]
national_female = national[national["Gender"].str.startswith("female")]
temp = ["National", "female"]
temp.append(national_female.sum()[2])
temp.append(national_female.sum()[5])
temp.append(national_female.sum()[6])
temp.append(round(float(national_female.sum()[7])/6, 4))
temp.append(round(float(national_female.sum()[5])/float(national_female.sum()[2]), 4))
temp.append(round(float(national_female.sum()[6])/float(national_female.sum()[2]), 4))
temp.append(round(float(national_female.sum()[7])/float(national_female.sum()[2]), 4))
df.loc[len(df)] = temp
national_male = national[national["Gender"].str.startswith("male")]
temp.clear()
temp = ["National", "male"]
temp.append(national_male.sum()[2])
temp.append(national_male.sum()[5])
temp.append(national_male.sum()[6])
temp.append(round(float(national_male.sum()[7])/6, 4))
temp.append(round(float(national_male.sum()[5])/float(national_male.sum()[2]), 4))
temp.append(round(float(national_male.sum()[6])/float(national_male.sum()[2]), 4))
temp.append(round(float(national_male.sum()[7])/float(national_male.sum()[2]), 4))
df.loc[len(df)] = temp
df
```
