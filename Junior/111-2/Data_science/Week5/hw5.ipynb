```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

r = requests.get("https://sys.ndhu.edu.tw/SA/XSL_ApplyRWD/ActApply.aspx?lang=tw") 
soup = BeautifulSoup(r.text, "html.parser")

act = []
hours = []
ppls = []
deadlines = []
dates = []

for i in range(0, 10):
    # print(i)
    name = soup.find("span", id = "BodyContent_gvActs_lblGv_act_name_"+str(i))
    hour = soup.find("span", id = "BodyContent_gvActs_lblGv_xsl_check_"+str(i))
    ppl = soup.find("span", id = "BodyContent_gvActs_lblGv_reg_num_"+str(i))
    deadline = soup.find("span", id = "BodyContent_gvActs_lblGv_reg_dt_"+str(i))
    date = soup.find("span", id = "BodyContent_gvActs_lblGv_act_dt_"+str(i))
    
    act.append(name.get_text())
    hours.append(hour.get_text()[4:])
    ppls.append(ppl.get_text().split(' ', 1)[0])
    deadlines.append(deadline.get_text())
    dates.append(date.get_text())

data = {
    "Name of the activity": act,
    "Number of hours certificated": hours,
    "Number of people who have registered": ppls,
    "Date and time of registration": deadlines,
    "Date and time of the activity": dates
}
# pd.set_option('display.unicode.ambiguous_as_wide', True)
# pd.set_option('display.unicode.east_asian_width', True)
# pd.set_option('display.width', 10000)
df = pd.DataFrame(data, index=range(1, 11))
df
# print(df)
```
