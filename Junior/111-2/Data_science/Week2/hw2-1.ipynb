Using loop to get factorial number from first n fibonacci number if first number is 1 and second number is 2.

Fibonacci example x1=1, x2=2, x3=x1+x2=3, x4=x2+x3=5

input 1

4

output 1 

1 factorial -> 1
2 factorial -> 2
3 factorial -> 6
5 factorial -> 120

input 2

6

output 2

1 factorial-> 1
2 factorial-> 2
3 factorial-> 6
5 factorial-> 120
8 factorial-> 40320
13 factorial-> 6227020800


```python
a=int(input())
num=1
FibArray = [1, 2]
#print(len(FibArray))
for i in range(1, a+1):
    if i==1 or i==2:
        k=FibArray[i-1]
    else:
        #print(i-1, i-2)
        #print(FibArray[i-2], FibArray[i-3])
        k=FibArray[i-2]+FibArray[i-3]
        FibArray.append(k)
    for j in range(1, k+1):
        num=num*j
    print(k,"factorial ->",num)
    num=1
        
```
