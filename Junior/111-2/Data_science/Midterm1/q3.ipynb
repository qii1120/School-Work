```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

r = requests.get("https://news.ycombinator.com")
soup = BeautifulSoup(r.text, "html.parser")

titles = []
scores = []
hrefs = []

title = soup.find_all("span", class_="titleline")
score = soup.find_all("td", class_="subtext")
# print(score[0])

for i in range(0, len(title)):
    hrefsoup = BeautifulSoup(str(title[i]), "html.parser")
    if hrefsoup.find("a").get('href')[:5] != "https":
        hrefs.append("https://news.ycombinator.com/" +
                     hrefsoup.find("a").get('href'))
    else:
        hrefs.append(hrefsoup.find("a").get('href'))
    t = title[i].get_text().split(" ")[len(title[i].get_text().split(" "))-1]
    if t[0] == '(':
        d = title[i].get_text().replace(t, "")
    else:
        d = title[i].get_text()
    scoresoup = BeautifulSoup(str(score[i]), "html.parser")
#     print(scoresoup.find("span", class_ = "score"))
#     print(i+1, d)
    titles.append(d)
    if scoresoup.find("span", class_="score"):
        scores.append(scoresoup.find("span", class_="score").get_text()[
                      :len(scoresoup.find("span", class_="score").get_text())-6])
    else:
        scores.append(scoresoup.find("span", class_="score"))

data = {
    "title": titles,
    "score": scores,
    "href": hrefs
}

df = pd.DataFrame(data, index=range(0, len(titles)))
df
# print(scores)
# print(len(hrefs))
# print(titles)
# print(r.text)
```
