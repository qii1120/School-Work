```python
import pandas as pd
import matplotlib.pyplot as plt

datas = pd.read_csv("data.csv")
plt.figure(figsize=(16, 16))
plt.subplot(2, 2, 1)
plt.plot(datas["Township"], datas["Male"], "b-", label="Male")
plt.plot(datas["Township"], datas["Female"], "r--", label="Female")
plt.plot(datas["Township"], datas["Total(Male+Female)"],
         "g-.", label="Total(Male+Female)")
plt.xticks(rotation=45)
plt.xlabel("Township")
plt.ylabel("number of people")
plt.legend()
plt.title("the line graphs of population")
plt.savefig("fig1.png", bbox_inches="tight")
plt.subplot(2, 2, 2)
plt.plot(datas["Township"], datas["Immigration"], "b-", label="Immigration")
plt.plot(datas["Township"], datas["Emigration"], "r--", label="Emigration")
plt.xticks(rotation=45)
plt.xlabel("Township")
plt.ylabel("number of people")
plt.legend()
plt.title("the line graphs of immigration and emigration")
plt.savefig("fig2.png", bbox_inches="tight")
plt.subplot(2, 2, 3)
plt.bar(datas["Township"], datas["Number_of_Households"])
plt.xticks(rotation=45)
plt.xlabel("Township")
plt.ylabel("number of people")
plt.title("the long bar graphs of households number")
plt.savefig("fig3.png", bbox_inches="tight")
plt.subplot(2, 2, 4)
plt.plot(datas["Township"], datas["Birth"], "b-", label="Birth")
plt.plot(datas["Township"], datas["Death"], "r--", label="Death")
plt.xticks(rotation=45)
plt.xlabel("Township")
plt.ylabel("number of people")
plt.legend()
plt.title("the line graphs of births and deaths")
plt.savefig("fig4.png", bbox_inches="tight")
plt.suptitle("Miaoli County population data")
plt.savefig("fig5.png", bbox_inches="tight")
plt.show()
# datas
```
